```python
import os
import nltk
import pandas as pd
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

## Data cleanning

```python
df = pd.read_csv("./Data/Source1/youtube_parsed_dataset.csv")
df
```

```python
df.drop(['index',"UserIndex","Number of Comments","Number of Subscribers","Membership Duration","Number of Uploads","Profanity in UserID","Age"], axis=1, inplace=True)
df
```

```python
df["Text"][1],df["oh_label"][1],
```

```python
def process_row(row):
    word_list = []
    for sent in nltk.sent_tokenize(row):
        word_list += nltk.word_tokenize(sent)
    return np.array([word for word in word_list if word not in nltk.corpus.stopwords.words('english')])
```

```python
df["Text"] = df["Text"].apply(process_row)
df["Text"]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
df = df.join(pd.DataFrame(mlb.fit_transform(df.pop('Text')),
                          columns=mlb.classes_,
                          index=df.index))
df
```
